Problem Statement 1:Smart Parking Lot Management System

```python
def smart_parking(capacity,logs):
 current=0
 peak=0
 for log in logs:
  if log=="IN":
   current+=1
  elif log=="OUT":
   current-=1
  if current>peak:
   peak=current
 print("Currently Parked Vehicles:",current)
 if current>capacity:
  print("Parking Status: Full / Over Capacity")
 else:
  print("Parking Status: Available")
 print("Peak Parking Usage:",peak)

capacity=50
logs=["IN","IN","IN","OUT","IN","IN","OUT"]
smart_parking(capacity,logs)
```

Problem Statement 2:Online Food Delivery Time Estimator

```python
def delivery_time(distance,traffic,weather):
 time=distance*5
 if traffic=="High":
  time+=10
 elif traffic=="Medium":
  time+=5
 if weather=="Rainy":
  time+=5
 elif weather=="Storm":
  time+=15
 print("Estimated Delivery Time:",time,"minutes")

delivery_time(8,"High","Rainy")
```

Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

```python
def seat_analysis(total,booked):
 count=len(booked)
 occupancy=(count/total)*100
 print("Occupancy:",int(occupancy),"%")
 if occupancy==100:
  print("Show Status: Housefull")
 elif occupancy>=70:
  print("Show Status: Almost Full")
 else:
  print("Show Status: Seats Available")
 if occupancy>90:
  print("Suggestion: Open Additional Show")

total=200
booked=[1]*150
seat_analysis(total,booked)
```

Problem Statement 4:Cloud Server Load Classification System

```python
def server_load(readings):
 total=0
 for r in readings:
  total+=r
 avg=total/len(readings)
 print("Average CPU Load:",int(avg),"%")
 if avg<50:
  print("Server Status: Normal")
 elif avg<=80:
  print("Server Status: Warning")
 else:
  print("Server Status: Critical")

readings=[45,60,70,85,90]
server_load(readings)
```

Problem Statement 5:Smart Classroom Resource Usage Monitor


```python
def resource_monitor(data):
 over=[]
 for k,v in data.items():
  if v>8:
   over.append(k)
 if over:
  print("Overused Resources:",",".join(over))
  print("Energy Alert: Yes")
 else:
  print("Overused Resources: None")
  print("Energy Alert: No")

data={"Projector":6,"AC":9,"Lights":4}
resource_monitor(data)
```

 Problem Statement 6: Online Event Registration Capacity Controller

```python
def event_registration(capacity,registrations):
 confirmed=0
 waitlist=0
 for i in range(registrations):
  if confirmed<capacity:
   confirmed+=1
  else:
   waitlist+=1
 print("Confirmed Registrations:",confirmed)
 print("Waitlisted Users:",waitlist)
 if waitlist>0:
  print("Registration Status: Closed")
 else:
  print("Registration Status: Open")

event_registration(100,105)
```

Problem Statement 1: Employee Performance Bonus Eligibility

```python
employees=eval(input("Enter employee dictionary: "))
max_score=max(employees.values())
top=[n for n,s in employees.items() if s==max_score]
print("Top Performers Eligible for Bonus:",",".join(top),f"(Score:{max_score})")
#INPUT:{"Ravi":92,"Anita":88,"Kiran":92,"Suresh":85}
```

Problem Statement 2: Search Query Keyword Analysis

```python
query=input().lower()
words=query.split()
freq={}
for w in words:
    freq[w]=freq.get(w,0)+1
result={k:v for k,v in freq.items() if v>1}
print(result)
#INPUT:Buy mobile phone buy phone online
```

Problem Statement 3: Sensor Data Validation

```python
sensor_readings=list(map(int,input().split(',')))
valid=[]
for i,v in enumerate(sensor_readings):
    if v%2==0:
        valid.append((i,v))
print("Valid Sensor Readings (Hour, Value):")
print(valid)
#INPUT:3, 4, 7, 8, 10, 12, 5
```

Problem Statement 4: Email Domain Usage Analysis

```python
emails=eval(input("Enter email list: "))
domain_count={}
for e in emails:
    domain=e.split("@")[1]
    domain_count[domain]=domain_count.get(domain,0)+1
total=len(emails)
for d,c in domain_count.items():
    percent=(c/total)*100
    print(f"{d}: {int(percent)}%")
#INPUT:["ravi@gmail.com","anita@yahoo.com","kiran@gmail.com","suresh@gmail.com","meena@yahoo.com"]
```

Problem Statement 5: Sales Spike Detection

```python
sales=eval(input())
avg=sum(sales)/len(sales)
threshold=round(avg*1.3)
for i,v in enumerate(sales,1):
    if v>=threshold:
        print(f"Day {i}: {v}")
#INPUT:[1200,1500,900,2200,1400,3000]
```

Problem Statement 6: Duplicate User ID Detection

```python
user_ids=eval(input())
freq={}
for u in user_ids:
    freq[u]=freq.get(u,0)+1
for k,v in freq.items():
    if v>1:
        print(f"{k} → {v} times")
#INPUT:["user1", "user2", "user1", "user3", "user1", "user3"]
```

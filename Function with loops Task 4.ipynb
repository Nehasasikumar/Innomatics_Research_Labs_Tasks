Problem 1: Mobile Recharge Validation System


```python
def validate_recharge(amount):
    valid_plans=[199,299,399,599]
    if amount<50:
        print("Recharge amount must be at least ₹50.")
        return False
    if amount not in valid_plans:
        print("Invalid recharge plan.")
        return False
    return True
while True:
    amount=int(input("Enter recharge amount: "))

    if validate_recharge(amount):
        print("Recharge successful!")
        break
    else:
        print("Please try again.\n")
#INPUT:299
```

Problem 2: Inventory Reorder Alert System

```python
def check_stock(products):
    for item,qty in products.items():
        if qty<15:
            print(item+": Reorder Alert")
        else:
            print(item+": Stock OK")
products={}
n=int(input("Enter number of products: "))
for i in range(n):
    name=input("Enter product name: ")
    qty=int(input("Enter quantity: "))
    products[name]=qty
check_stock(products)
''' 
INPUT:
Enter number of products: 3
Enter product name: Rice
Enter quantity: 20
Enter product name: Sugar
Enter quantity: 10
Enter product name: Oil
Enter quantity: 5
'''
```

Problem 3: Student Result Processing System


```python
def result_status(marks):
    total=0
    for m in marks:
        total+=m
    avg=total/len(marks)

    if avg>=50:
        return "Pass"
    else:
        return "Fail"
marks=[]
n=int(input("Enter number of subjects: "))
for i in range(n):
    m=int(input("Enter marks: "))
    marks.append(m)
print(result_status(marks))
'''
Enter number of subjects: 3
Enter marks: 60
Enter marks: 40
Enter marks: 50
'''
```

Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance,peak):
    base=50
    per_km=12
    fare=base+(per_km*distance)
    if peak.lower()=="yes":
        fare=fare+(fare*0.25)
    return fare
while True:
    distance=float(input("Enter distance in km: "))
    peak=input("Is it peak hour? (yes/no): ").strip()

    total=calculate_fare(distance,peak)
    print("Total Fare: ₹",total)

    retry=input("Do you want to calculate again? (yes/no): ")
    if retry.lower()!="yes":
        break
'''
Enter distance in km: 10
Is it peak hour? (yes/no): yes
Do you want to calculate again? (yes/no): no
'''
```

Problem 5: Employee Attendance Eligibility Checker


```python
def check_eligibility(attendance):
    present=0
    for i in attendance:
        if i=="P":
            present+=1
    percent=(present/len(attendance))*100

    if percent>=75:
        return "Eligible"
    else:
        return "Not Eligible"
attendance=[]
n=int(input("Enter number of days: "))
for i in range(n):
    status=input("Enter P/A: ").upper()
    attendance.append(status)
print(check_eligibility(attendance))
'''
Enter number of days: 4
P
P
A
P
'''
```

Problem 6: Password Strength Checker

```python
def check_password(pwd):
    digit=False
    special=False
    if len(pwd)<8:
        return "Weak Password"
    for ch in pwd:
        if ch.isdigit():
            digit=True
        if ch in "@#$":
            special=True
    if digit and special:
        return "Strong Password"
    else:
        return "Weak Password"
pwd=input("Enter password: ")
print(check_password(pwd))
#INPUT:abc123@
```

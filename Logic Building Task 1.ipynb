User Login Check

```python
username=input("Enter username: ")
password=input("Enter password: ")
if username=='admin' and password=='1234':
    print("Login Successful")
else:
    print("Invalid Credentials")
```

Pass / Fail Analyzer

```python
marks=eval(input("Enter the list of marks: "))
pass_count=0
fail_count=0
for mark in marks:
    if mark>=50:
        pass_count+=1
    else:
        fail_count+=1
print("Total number of students:",len(marks))
print("Pass Students:",pass_count)
print("Fail Students:",fail_count)
```

Simple Data Cleaner

```python
names=eval(input("Enter the datas: "))
cleaned_names=[name.strip().lower() for name in names]
print("Cleaned Datas : ",cleaned_names)
```

Message Length Analyzer

```python
messages=eval(input("Enter the messages: "))
flag=0
for message in messages:
    length=len(message)
    print("Length:",length)
    if length>10:
        flag+=1
print("Flagged messages:",flag)
```

Error Message Detector

```python
logs=eval(input("Enter logs:"))
error_count=logs.count("ERROR")
print("Total ERROR entries:",error_count)
```

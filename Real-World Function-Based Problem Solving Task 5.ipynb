Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_engagement(likes):
    total_likes=0
    for i in likes:
        total_likes+=i
    print("Total Likes:",total_likes)
    if total_likes>=1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")
likes=list(map(int,input().split()))
analyze_engagement(likes)
#INPUT: 1 2 500 8000 20
```

Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def rainfall_checker(rain,req):
    total=0
    for i in rain:
        total+=i
    avg=total//len(rain)
    print("Average Rainfall:",avg)
    if avg>=req:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")
rain=list(map(int,input().split()))
req=int(input())
rainfall_checker(rain,req)
#INPUT: 30 6
#INPUT: 30
```

Problem 4: Social Media – Duplicate Account Detection

```python
def check_duplicates(users):
    if len(users)!=len(set(users)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
n=int(input())
users=[]
for i in range(n):
    users.append(input())
check_duplicates(users)
'''
INPUT: 5
john
emma
alex
john
mike'''
```

Problem 6: Agriculture – Premium Crop Price Filter


```python
def premium_crop_filter(prices):
    premium=[]
    for i in prices:
        if i>2000:
            premium.append(i)
    print("Premium Crops:",premium)
prices=list(map(int,input().split()))
premium_crop_filter(prices)
#INPUT: 1500 2500 1800 3200 900
```

Problem 7: System Monitoring – Application Health Checker

```python
def check_health(errors):
    print("Error Count:",errors)
    if errors==0:
        print("System Status: Healthy")
    elif errors<=5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")
errors=int(input())
check_health(errors)
```

Problem 9: E-Learning – Student Attendance Eligibility System

```python
def check_attendance(attendance):
    total_classes=len(attendance)
    present=0
    for i in attendance:
        if i==1:
            present+=1
    percentage=(present/total_classes)*100
    print("Attendance Percentage:",percentage)
    if percentage>=75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")
attendance=list(map(int,input().split()))
check_attendance(attendance)
#INPUT: 1 1 1 0 1
```

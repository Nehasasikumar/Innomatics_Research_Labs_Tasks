Problem 1: Smart Light Controller

```python
class SmartLightController:
 def __init__(self,name):
  self.name=name
  self.status="OFF"
 def turn_on(self):
  self.status="ON"
 def turn_off(self):
  self.status="OFF"
 def show_status(self):
  print(self.name,"is",self.status)
name=input("Light Name: ")
action=input("Action: ")
light=SmartLightController(name)
if action=="ON":
 light.turn_on()
elif action=="OFF":
 light.turn_off()
light.show_status()
'''INPUT: Light Name: Bedroom Light
Action: ON'''
```

Problem 2: Employee ID Card System

```python
class Employee:
 def __init__(self,name,id,dept):
  self.name=name
  self.id=id
  self.dept=dept
 def display(self):
  print("Employee ID Card")
  print("Name:",self.name)
  print("ID:",self.id)
  print("Department:",self.dept)
name=input("Employee Name: ")
id=input("Employee ID: ")
dept=input("Department: ")
emp=Employee(name,id,dept)
emp.display()
'''INPUT:
Rahul
EMP102
IT'''
```

Problem 3: Mobile Contact Record

```python
class Contact:
 def __init__(self,name,phone):
  self.name=name
  self.phone=phone
 def display(self):
  print("Contact Saved")
  print("Name:",self.name)
  print("Phone:",self.phone)
name=input("Contact Name: ")
phone=input("Phone Number: ")
c=Contact(name,phone)
c.display()
'''INPUT:
Anita
9876543210'''
```

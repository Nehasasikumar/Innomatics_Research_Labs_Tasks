Problem 1: Unique Words in a Sentence with case sensitivity

```python
sentence=input("Enter sentence:")
words=sentence.split()
unique_words=set(words)
print("Unique words count:",len(unique_words))
print("Unique words:",unique_words)
#INPUT: Python is powerful and python is easy
```

Problem 1: Unique Words in a Sentence without case sensitivity

```python
sentence=input("Enter sentence:")
words=sentence.lower().split()
unique_words=set(words)
print("Unique words count:",len(unique_words))
print("Unique words:",unique_words)
#INPUT: Python is powerful and python is easy
```

Problem 2: Highest Salary from Employee Data

```python
employees={}
n=int(input("Enter number of employees:"))
for i in range(n):
    name=input("Enter employee name:")
    salary=int(input("Enter salary:"))
    employees[name]=salary
highest_employee=max(employees,key=employees.get)
highest_salary=employees[highest_employee]
print("Highest Salary:",highest_employee,"-",highest_salary)
#INPUT
#Enter number of employees:3
#Enter employee name:Ravi
#Enter salary:8000
#Enter employee name:Ram
#Enter salary:10000
#Enter employee name:Mani
#Enter salary:5000
```

Problem 3: Find Maximum and Minimum Values

```python
numbers=list(map(int,input().split()))
max_val=numbers[0]
min_val=numbers[0]
for num in numbers:
    if num>max_val:
        max_val=num
    if num<min_val:
        min_val=num
print("List:",numbers)
print("Max:",max_val)
print("Min:",min_val)
#INPUT:1 2 3 4 5
```

Problem 4: Count Products Above a Price Threshold

```python
prices=list(map(int,input().split()))
count=0
for price in prices:
    if price>1000:
        count+=1
print("Products above 1000:",count)
#INPUT:1000 50 1200 900 20
```

Problem 5: Calculate Attendance Percentage

```python
attendance=list(map(str,input().upper().split()))
present_count=attendance.count("P")
percentage=(present_count/len(attendance))*100
print("Attendance Percentage:",percentage)
#INPUT:A P A P A P
```

Problem 7: Count Character Frequency

```python
text=input("Enter a string:")
freq={}
for ch in text:
    if ch in freq:
        freq[ch]+=1
    else:
        freq[ch]=1
print(freq)
#INPUT:Python
```
